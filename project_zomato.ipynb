# Cleaning Data

```python
# Checking for duplicated rows
zomato.duplicated()\
        .sum()
```

```python
# Removing all duplicates from the dataset
zomato.drop_duplicates(keep=False, inplace=True)

# Looking at the dataset after changes
zomato.duplicated()\
        .sum()
```

```python
# Removing all NaN values from the dataset
zomato.dropna(how='any', inplace=True)

# Looking at the dataset after changes
zomato.isna()\
        .sum()\
        .sort_values(ascending=False)
```

```python
zomato.columns
```

```python
# Renaming the columns for readability
zomato = zomato.rename(columns={'approx_cost(for two people)': 'cost',
                       'listed_in(type)': 'type',
                       'listed_in(city)': 'city'})

# Looking at the dataset after changes
zomato.columns
```

```python
# Let's see what the unique data looks like in the 'cost' column
zomato.cost\
        .unique()
```

```python
# Converting all string values ​​in the 'cost' column to float values
zomato.cost = zomato.cost\
                    .apply(lambda x: x.replace(',', '') if type(x) == str else x)\
                    .astype(float)

# Looking at the dataset after changes
zomato.cost\
        .unique()
```

```python
# Let's see what the unique data looks like in the 'rate' column
zomato.rate\
        .unique()
```

```python
# Getting rid of string values ​​in the 'rate' column
zomato = zomato.loc[zomato.rate != 'NEW']
zomato = zomato.loc[zomato.rate != '-']

# Converting of all string values ​​in the 'rate' column to float values
zomato.rate = zomato.rate\
                    .apply(lambda x: x.replace('/5', '') if type(x) == str else x)\
                    .str\
                    .strip()\
                    .astype(float)
```

```python
# Looking at the dataset after changes
zomato.rate\
        .unique()
```

```python
# Changing the values in the "online_order" column
zomato.online_order\
        .replace(('Yes','No'), (1, 0), inplace=True)

# Changing the values in the "book_table" column
zomato.book_table\
        .replace(('Yes','No'), (1, 0), inplace=True)

# Looking at the dataset after changes
zomato.head()
```

# Data Analysis

### Finding average check by converting the average check into the current USD rate and the current EUR rate

```python
# Importing libraries for scraping
import requests
from bs4 import BeautifulSoup
```

```python
# The current exchange rate of the Indian rupee against the USD and the EUR (forbes is a good resource to trust)
url_rupi_dollar = 'https://www.forbes.com/advisor/money-transfer/currency-converter/inr-usd/'
url_rupi_euro = 'https://www.forbes.com/advisor/money-transfer/currency-converter/inr-eur/'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:104.0) Gecko/20100101 Firefox/104.0'}
```

```python
# Getting the page and scraping
page_rupi_dollar = requests.get(url_rupi_dollar, headers=headers)
page_rupi_euro = requests.get(url_rupi_euro, headers=headers)

soup_rupi_dollar = BeautifulSoup(page_rupi_dollar.content, 'html.parser')
soup_rupi_euro = BeautifulSoup(page_rupi_euro.content, 'html.parser')
```

```python
# Getting currency rate values
convert_dollar = soup_rupi_dollar.find_all('span', {'class': 'amount'})
convert_euro = soup_rupi_euro.find_all('span', {'class': 'amount'})
```

```python
dollar = float(convert_dollar[0].text)  # The current exchange rate to the USD
euro = float(convert_euro[0].text)  # The current exchange rate to the EUR
```

```python
# The unique values ​​of the 'cost' column (check cost)
zomato.cost.unique()
```

```python
# Selection of restaurants with an existing ticket
zomato_exchange = zomato.query('cost > 0')
```

```python
# Creating new columns and convert the check to the current USD and EUR rates
zomato_exchange['pay_usd'] = round(zomato_exchange.cost * dollar, 2)
zomato_exchange['pay_eur'] = round(zomato_exchange.cost * euro, 2)
```

### Finding the most popular restaurants in Bangalore

```python
# Importing libraries for visualizing data
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import seaborn as sns

# Setting figure size for future visualizations
sns.set(rc={'figure.figsize':(10,8)})
sns.set_style('whitegrid')
```

```python
# First, let's see what famous reataurant chains are in Bangalore
chains = zomato['name']\
    .value_counts()[:20]

sns.barplot(x=chains
            , y=chains.index
            , palette='Set1')

plt.title("Most famous restaurant chains in Bangaluru"
          , size=20
          , pad=20)
plt.xlabel("Number of outlets"
           , size=15)

sns.despine()
```

```python
# Let's see at the cost distribution of all the restaurants in the city
sns.histplot(zomato.cost
            , stat='percent'
            , palette='Set1'
            , bins=30)

plt.title("Cost Distrubution for all Restaurants"
          , size=20
          , pad=20)

sns.despine()
```
