```python
# Importing the kaggle library to load a dataset
import kaggle
from kaggle.api.kaggle_api_extended import KaggleApi
```

```python
# Authentificating 
api = KaggleApi()
api.authenticate()

# And load a dataset
api.dataset_download_file('himanshupoddar/zomato-bangalore-restaurants',
                         file_name='zomato.csv')
```
